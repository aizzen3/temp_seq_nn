```python
import pandas as pd
import numpy as np

df = pd.read_csv("temperature_signals.csv")

# take corrupted signals (this is what model sees)
X = df[["signal1_corrupted", "signal2_corrupted"]].values
```

```python
# 1 = observed, 0 = missing
observed_mask = ~np.isnan(X)

print("Original missing values:", np.isnan(X).sum())
```

```python
missing_rate = 0.2  # 20% masking

# random mask
random_mask = np.random.rand(*X.shape) < missing_rate

# only apply masking where data is already present
artificial_mask = random_mask & observed_mask
```

```python
X_masked = X.copy()

# hide values
X_masked[artificial_mask] = np.nan
```

```python
# this tells model what is visible
final_mask = ~np.isnan(X_masked)

# this tells where model should be evaluated
target_mask = artificial_mask
```

```python
import matplotlib.pyplot as plt

t = df["time"].values

plt.figure(figsize=(10,4))
plt.plot(t, X[:,0], label="Original Signal")
plt.scatter(t[target_mask[:,0]], X[target_mask[:,0],0],
            color='red', label="Masked Points")
plt.legend()
plt.title("Masking Visualization (Signal 1)")
plt.show()
```
